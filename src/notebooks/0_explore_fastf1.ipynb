### Exploring the fastf1 Library

```python
import fastf1
```

```python
last_year_events = fastf1.get_event_schedule(2024)
last_year_events
```

```python
shanghai_sq_last_year = fastf1.get_session(2024, "Shanghai", "SQ")
shanghai_sq_last_year.load()
shanghai_sq_last_year.results
```

```python
shanghai_s_last_year = fastf1.get_session(2024, "Shanghai", "Sprint")
shanghai_s_last_year.load()
shanghai_s_last_year.results
```

```python
def find_sprint_races(schedule):
    sprint_schedules = schedule[schedule['EventFormat'].str.contains('sprint', case=False, na=False)]
    return sprint_schedules

sprints_last_year = find_sprint_races(last_year_events)
sprints_last_year
```
